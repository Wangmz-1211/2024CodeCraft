```python
import warnings

warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
import numpy as np
```

```python
def show_result(path):
    plt.figure(figsize=(12,6))
    with open(path, 'r') as f:
        lines = f.readlines()
    history = []
    for i in range(len(lines)):
        line = lines[i].strip()
        if i % 3 == 0:
            arg1 = int(line.split(' ')[0])
            arg2 = int(line.split(' ')[1])
            history.append([arg1, arg2])
        if i % 3 == 1:
            res = json.loads(line)
            history[-1].append(int(res['score']))
    history = np.array(history)
    df = pd.DataFrame(history, columns=['H_ASTAR_MAX_TIME', 'H_DOCK_PUNISH', 'score'])
    sns.heatmap(df.pivot_table(index='H_ASTAR_MAX_TIME', columns='H_DOCK_PUNISH', values='score'), annot=True,
                fmt=".1f", cmap='viridis')
```

# Grid Search Results in Heatmap 

```python
show_result('morethan15/map-3.10.history')
plt.title('map-3.10')
```

```python
show_result('morethan15/map-3.8.history.1')
plt.title('map-3.8')
```

```python
show_result('morethan15/map-3.8.history.2')
plt.title('map-3.8')
```

```python
show_result('morethan15/map-3.8.history.3')
plt.title('map-3.8')
```
